```python
import pandas as pd
```

```python
df = pd.DataFrame(
    {
      "dates": [
         pd.Timestamp("2021-06-01"),
         pd.Timestamp("2021-07-11"),
         pd.Timestamp("2021-06-25"),
         pd.Timestamp("2021-06-29"),
         pd.Timestamp("2021-03-20"),
         pd.Timestamp("2021-01-22"),
         pd.Timestamp("2021-01-06"),
         pd.Timestamp("2020-01-01"),
         pd.Timestamp("2021-05-09")
      ],

      "Ratings": ["A", "B", "A", "B","N", "G", "D", "H", "E"]
   }
)
```

## Extract Different Parts from dates

```python
df['day']=df['dates'].dt.day
df['dayofweek']=df['dates'].dt.dayofweek
df['day_of_year']=df['dates'].dt.day_of_year

df['weekofyear']=df['dates'].dt.isocalendar().week


df['month']=df['dates'].dt.month

df['quarter']=df['dates'].dt.quarter

df['year']=df['dates'].dt.year

df['is_month_start']=df['dates'].dt.is_month_start
df['is_month_end']=df['dates'].dt.is_month_end



df['is_year_start']=df['dates'].dt.is_year_start
df['is_leap_year']=df['dates'].dt.is_leap_year




df
```

## Difference between 2 dates in different Units (days,weeks,months,year ,hours,minutes,seconds)

```python
a = pd.date_range(start="2019-12-29", freq="d", periods=10)
b = pd.date_range(start="2020-02-14", freq="d", periods=10)
a,b
```

```python
df=pd.DataFrame(data=(a,b))
df
```

```python
df=pd.DataFrame({'a':a,'b':b})
df
```

```python
df.dtypes
```

```python
import numpy as np


df['diff_days'] = (df['b'] - df['a']) / np.timedelta64(1, 'D')
df['diff_weeks'] = (df['b'] - df['a']) / np.timedelta64(1, 'W')
df['diff_months'] = (df['b'] - df['a']) / np.timedelta64(1, 'M')
df['diff_years'] = (df['b'] - df['a']) / np.timedelta64(1, 'Y')
df
```

```python
df = pd.DataFrame({'start_date': ['2020-01-05', '2020-01-12', '2020-01-19'],
                   'end_date': ['2020-06-30', '2020-07-31', '2020-08-31']})

#view dtype of each column
print(df.dtypes)
```

```python
#convert columns to datetime
df[['start_date','end_date']] = df[['start_date','end_date']].apply(pd.to_datetime)

#calculate difference between dates
df['diff_days'] = (df['end_date'] - df['start_date']) / np.timedelta64(1, 'D')

#view updated DataFrame
print(df)
```
