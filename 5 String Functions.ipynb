````python
df1=pd.DataFrame()

df1['name']=df['Name']
df1['lower']=df['Name'].str.lower() 
df1['upper']=df['Name'].str.upper()
df1['capitalize']=df['Name'].str.capitalize()
df1['title']=df['Name'].str.title()
df1['casefold']=df['Name'].str.casefold()
df1['swapcase']=df['Name'].str.swapcase()

df1
````

````python
df2=pd.DataFrame()

df2['name']=df['Name']
df2['concat']=df['Sex'].str.cat(df['Name'],sep=' -$- ')
df2['R-count']=df['Name'].str.count('r')
df2['length']=df['Name'].str.len()
df2['contains']=df['Name'].str.contains('mr', case=False, flags=0, na=None, regex=False)
df2['extract']=df['Name'].str.extract(r'(\D{2,3})?')
df2['center_Padding']=df['Name'].str.center(28, fillchar='#')
df2['right_Padding']=df['Name'].str.ljust(28, fillchar='#')
df2
````

````python
df3=pd.DataFrame()
df3=df
df3['find-r']=df['Name'].str.find('r') # same as #df3['index']=df['Name'].str.index('r') , index() raises exception whereas find() returns -1 if substring is not found
df3['find r-after 3rd char']=df['Name'].str.find('r',start=3)
df3['findall-ra']=df['Name'].str.findall('ra') #pattern Match
df3['get nth Index char']=df['Name'].str.get(15)
df3['strip']=df['Name'].str.strip(',.')
df3['repeat']=df['Sex'].str.repeat(repeats=3)
df3['replace']=df['Sex'].str.replace('male','m')
df3
````
