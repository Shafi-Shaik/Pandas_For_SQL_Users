##  SELECT with WHERE clause

```
SELECT *
FROM exercise.titanic 
WHERE passengerid =2
```

````python
titanic[titanic['passengerid']==2] # or titanic.loc[titanic['passengerid']==2]
````

```
SELECT passengerid,name,sex
FROM exercise.titanic 
WHERE passengerid =2
```

````python
titanic.loc[titanic['passengerid']==2,['passengerid','name','sex']]
````

##  SELECT - WHERE - IN clauses

```
SELECT passengerid,name,sex
FROM exercise.titanic 
WHERE passengerid in (2 ,35,78)
```

````python
titanic.loc[titanic['passengerid'].isin([2,35,78]),['passengerid','name','sex']]
````

##  SELECT - WHERE - OR clauses

```
SELECT passengerid,name,sex
FROM exercise.titanic 
WHERE passengerid =2 
	OR  sex ='male'
```

````python
titanic.loc[(titanic['passengerid'].isin([2])) | (titanic['sex']=='female'),['passengerid','name','sex']]
````

##  SELECT - WHERE - AND clauses

```
SELECT passengerid,name,sex
FROM exercise.titanic 
WHERE passengerid in (2,35,67)
	 AND sex ='female'
```

````python
titanic.loc[(titanic['passengerid'].isin([2,35,67])) & (titanic['sex']=='female'),['passengerid','name','sex']]
````

##  SELECT - WHERE - BETWEEN clauses

```
SELECT passengerid,name,age
FROM exercise.titanic 
WHERE passengerid in (2,35,67)
	 AND age between 20 and 29
```

````python
titanic.loc[(titanic['passengerid'].isin([2,35,67])) & (titanic['age'].between(20,29)),['passengerid','name','age']]
````

##  SELECT - WHERE - NULL clauses

```
SELECT passengerid,name,age
FROM exercise.titanic 
WHERE age is  null
```

````python
titanic.loc[titanic['age'].isnull(),['passengerid','name','age']]
````

##  SELECT - WHERE - NOT NULL clauses

```
SELECT passengerid,name,age
FROM exercise.titanic 
WHERE age is not null
```

````python
titanic.loc[~titanic['age'].isnull(),['passengerid','name','age']]
````

##  SELECT - WHERE - LIKE clauses

```
SELECT passengerid,name,age
FROM exercise.titanic 
WHERE name like '%sha%'
```

````python
titanic.loc[titanic['name'].str.contains('sha|Sha',regex=True, na=True),['passengerid','name','age']]
````

```
SELECT passengerid,name,age
FROM exercise.titanic 
WHERE name like 'sha%'
```

````python
titanic.loc[titanic['name'].str.lower().str.startswith('sha'),['passengerid','name','age']]
````

```
SELECT passengerid,name,age
FROM exercise.titanic 
WHERE name like '%sh'
```

````python
titanic.loc[titanic['name'].str.lower().str.endswith('sh'),['passengerid','name','age']]
````
